```python
def load_in_data(input_file, xarray_name, start_year, end_year, lat_lbound, lat_ubound):
    
    paths = glob(input_file)
    year_limited_paths = limit_files_to_select_years(paths,range(start_year,end_year+1))
    dataset = xr.open_mfdataset(year_limited_paths, combine="by_coords")
    dataset = dataset.sortby('lat',ascending=True)

    data_var = dataset[xarray_name].sel(time = slice(str(start_year)+'-01-01', str(end_year)+'-12-31'), lat = slice(lat_lbound,lat_ubound))
    data_var.load()

    return dataset, data_var
```

```python
def remove_seasonal_cycle(data_var,sample_rate):
    fft_result = np.fft.rfft(data_var, axis=0)

    N = len(data_var)
    
    freq = np.fft.rfftfreq(N, d=1/sample_rate)

    # identify the three harmonics related to the seasonal cycle
    seasonal_frequencies = [1/365.25, 2/365.25, 3/365.25]  # unit cycles per day
    harmonic_indices = [np.argmin(np.abs(freq - f)) for f in seasonal_frequencies]
    
    # create a fft varaible with just the seasonal cycle components - mean and first three harmonics

    fft_seasonal = np.zeros(fft_result.shape, dtype = fft_result.dtype)
    fft_seasonal[0] = fft_result[0]
    fft_seasonal[harmonic_indices] = fft_result[harmonic_indices]

    seasonal_cycle = np.fft.irfft(fft_seasonal, n=N, axis = 0)

    # create xarray DataArray for seasonal cycle

    full_lat = data_var.lat
    full_lon = data_var.lon
    full_time = data_var.time
    seasonal_cycle = xr.DataArray(data = seasonal_cycle, dims=["time","lat","lon"], coords=[("time",full_time.data),("lat",full_lat.data),("lon",full_lon.data)])

    #remove seasonal cycle from original data
    
    data_var_ns = data_var - seasonal_cycle


    return seasonal_cycle, data_var_ns
```

```python
def remove_seasonal_cycle_2(data_var):

    seasonal_cycle = data_var.groupby('time.dayofyear').mean(dim='time')

    fft_result = np.fft.rfft(seasonal_cycle, axis = 0)
    N = len(seasonal_cycle)

    #selecting the mean and first three harmonics
    harmonic_indices = [0, 1, 2, 3]
    fft_seasonal = np.zeros(fft_result.shape, dtype = fft_result.dtype)
    fft_seasonal[harmonic_indices] = fft_result[harmonic_indices]

    seasonal_cycle_tem = np.fft.irfft(fft_seasonal, n=N, axis = 0)

    seasonal_cycle_smooth = seasonal_cycle.copy(data = seasonal_cycle_tem)

    data_var_ns = data_var.groupby('time.dayofyear') - seasonal_cycle_smooth

    return seasonal_cycle_smooth, data_var_ns
```

```python
def save_output_monthly_chunks(start_year, end_year, var_data, output_dir, output_name):

    var_data = xr.DataArray(var_data, name = output_name)

    for year in range(start_year, end_year + 1):
    
        for month in range(1,13):
        
            if month == 12:
                tem = var_data.sel(time = slice(str(year)+'-'+str(month).zfill(2)+'-01',str(year)+'-'+str(month).zfill(2)+'-31'))

            else:
                last_day_of_month = pd.Timestamp(f'{str(year)}-{str(month+1).zfill(2)}-01') - pd.Timedelta(days=0.5)
                tem = var_data.sel(time = slice(str(year)+'-'+str(month).zfill(2)+'-01',last_day_of_month))


            output_file = output_dir+output_name+'.'+str(year)+str(month).zfill(2)+'.nc'
            tem.to_netcdf(output_file)
```
