```python
# Define constants

g = 9.8 #[m s^-2]
L = 2.26e6 #[J/kg]
cp = 1005 #[J/kg-K]
R_e = 6.378e6 #[m]
pi = 22/7

#########################################
# Define paths of files we wish to load #
#########################################
    
# glob expands paths with * to a list of files, like the unix shell #

paths_MSE = glob(input_file_MSE)
paths_col_MSE = glob(input_file_col_MSE)
paths_VADV = glob(input_file_VADV)
paths_col_VADV = glob(input_file_col_VADV)
paths_HADV = glob(input_file_HADV)
paths_col_HADV = glob(input_file_col_HADV)
paths_HADV_zonal = glob(input_file_HADV_zonal)
paths_col_HADV_zonal = glob(input_file_col_HADV_zonal)
paths_HADV_meridional = glob(input_file_HADV_meridional)
paths_col_HADV_meridional = glob(input_file_col_HADV_meridional)

paths_jim_col_budget = glob(input_file_jim_col_budget)
paths_jim_VADV = glob(input_file_jim_VADV)
paths_jim_HADV = glob(input_file_jim_HADV)

# Limit paths #

years_limited_paths_MSE = limit_files_to_select_years(paths_MSE,range(start_year,end_year+1))
years_limited_paths_col_MSE = limit_files_to_select_years(paths_col_MSE,range(start_year,end_year+1))
years_limited_paths_VADV = limit_files_to_select_years(paths_VADV,range(start_year,end_year+1))
years_limited_paths_col_VADV = limit_files_to_select_years(paths_col_VADV,range(start_year,end_year+1))
years_limited_paths_HADV = limit_files_to_select_years(paths_HADV,range(start_year,end_year+1))
years_limited_paths_col_HADV = limit_files_to_select_years(paths_col_HADV,range(start_year,end_year+1))
years_limited_paths_HADV_zonal = limit_files_to_select_years(paths_HADV_zonal,range(start_year,end_year+1))
years_limited_paths_col_HADV_zonal = limit_files_to_select_years(paths_col_HADV_zonal,range(start_year,end_year+1))
years_limited_paths_HADV_meridional = limit_files_to_select_years(paths_HADV_meridional,range(start_year,end_year+1))
years_limited_paths_col_HADV_meridional = limit_files_to_select_years(paths_col_HADV_meridional,range(start_year,end_year+1))

years_limited_paths_jim_col_budget = limit_files_to_select_years(paths_jim_col_budget,range(start_year,end_year+1))
years_limited_paths_jim_VADV = limit_files_to_select_years(paths_jim_VADV,range(start_year,end_year+1))
years_limited_paths_jim_HADV = limit_files_to_select_years(paths_jim_HADV,range(start_year,end_year+1))

print(len(years_limited_paths_jim_col_budget))

##### LAZY LOADING in DATA #####

dataset_MSE = xr.open_mfdataset(years_limited_paths_MSE, combine="by_coords")
dataset_col_MSE = xr.open_mfdataset(years_limited_paths_col_MSE, combine="by_coords")
dataset_VADV = xr.open_mfdataset(years_limited_paths_VADV, combine="by_coords")
dataset_col_VADV = xr.open_mfdataset(years_limited_paths_col_VADV, combine="by_coords")
dataset_HADV = xr.open_mfdataset(years_limited_paths_HADV, combine="by_coords")
dataset_col_HADV = xr.open_mfdataset(years_limited_paths_col_HADV, combine="by_coords")
dataset_HADV_zonal = xr.open_mfdataset(years_limited_paths_HADV_zonal, combine="by_coords")
dataset_col_HADV_zonal = xr.open_mfdataset(years_limited_paths_col_HADV_zonal, combine="by_coords")
dataset_HADV_meridional = xr.open_mfdataset(years_limited_paths_HADV_meridional, combine="by_coords")
dataset_col_HADV_meridional = xr.open_mfdataset(years_limited_paths_col_HADV_meridional, combine="by_coords")

dataset_jim_col_budget = xr.open_mfdataset(years_limited_paths_jim_col_budget, combine="by_coords")
dataset_jim_VADV = xr.open_mfdataset(years_limited_paths_jim_VADV, combine="by_coords")
dataset_jim_HADV = xr.open_mfdataset(years_limited_paths_jim_HADV, combine="by_coords")

#### LOAD DATA ####

full_lat = dataset_jim_col_budget['lat']
full_lon = dataset_jim_col_budget['lon']

col_VADV_jim = dataset_jim_col_budget['viWDMDP'].sel(time = slice(str(start_year)+'-01-01', str(end_year)+'-12-31'), lat = slice(-15,15))
col_HADV_jim = dataset_jim_col_budget['viVDOTDELM'].sel(time = slice(str(start_year)+'-01-01', str(end_year)+'-12-31'), lat = slice(-15,15))
precip_jim = dataset_jim_col_budget['PRECT'].sel(time = slice(str(start_year)+'-01-01', str(end_year)+'-12-31'), lat = slice(-15,15))

col_VADV_vijit = dataset_col_VADV['__xarray_dataarray_variable__']
col_HADV_vijit = dataset_col_HADV['__xarray_dataarray_variable__']

col_VADV_jim.load()
col_HADV_jim.load()
precip_jim.load()
col_VADV_vijit.load()
col_HADV_vijit.load()
```

```python
#### Scatter and correlation at some sample locations in Indian Ocean and Pacific ocean ###

lat_tem = 0
lon_tem = 75

data_tem_vijit = col_VADV_vijit.sel(lat = lat_tem, lon = lon_tem)
data_tem_jim = col_VADV_jim.sel(lat = lat_tem, lon = lon_tem)
data_tem_prc = precip_jim.sel(lat = lat_tem, lon = lon_tem)

plt.scatter(data_tem_jim, data_tem_vijit)
plt.xlabel('VADV jim')
plt.ylabel('VADV vijit')
plt.title('VADV comparison')

slope, intercept = np.polyfit(data_tem_jim, data_tem_vijit, 1)
plt.plot(data_tem_jim, slope * data_tem_jim + intercept, color='red', label=f'y = {slope:.2f}x + {intercept:.2f}')
plt.plot(data_tem_jim,data_tem_jim,color='grey', label='1:1 line')
plt.legend()

plt.show()

correlation = np.corrcoef(data_tem_jim, data_tem_vijit)[0, 1]
print(f'The correlation is {correlation:.2f}.')

data_tem_vijit = col_HADV_vijit.sel(lat = lat_tem, lon = lon_tem)
data_tem_jim = col_HADV_jim.sel(lat = lat_tem, lon = lon_tem)

plt.scatter(data_tem_jim, data_tem_vijit)
plt.xlabel('HADV jim')
plt.ylabel('HADV vijit')
plt.title('HADV comparison')

slope, intercept = np.polyfit(data_tem_jim, data_tem_vijit, 1)
plt.plot(data_tem_jim, slope * data_tem_jim + intercept, color='red', label=f'y = {slope:.2f}x + {intercept:.2f}')
plt.plot(data_tem_jim,data_tem_jim,color='grey', label='1:1 line')
plt.legend()

plt.show()

correlation = np.corrcoef(data_tem_jim, data_tem_vijit)[0, 1]
print(f'The correlation is {correlation:.2f}.')

plt.plot(data_tem_prc,color='black',label='precip')
#plt.plot(data_tem_jim,color='blue',label='data')
plt.legend()
plt.show()
```

```python
def mean_max_min_spatial_map(data):

    time_mean = data.mean(dim='time')
    time_max = data.max(dim='time')
    time_min = data.min(dim='time')

    # Plot the spatial maps
    fig, axs = plt.subplots(1, 3, figsize=(15, 5))
    time_mean.plot(ax=axs[0], cmap='coolwarm', add_colorbar=False)
    axs[0].set_title('Time Mean')
    time_max.plot(ax=axs[1], cmap='coolwarm', add_colorbar=False)
    axs[1].set_title('Maximum')
    time_min.plot(ax=axs[2], cmap='coolwarm', add_colorbar=False)
    axs[2].set_title('Minimum')
    fig.colorbar(axs[2].collections[0], ax=axs, location='right', shrink=0.6)

    plt.show()
    
```
